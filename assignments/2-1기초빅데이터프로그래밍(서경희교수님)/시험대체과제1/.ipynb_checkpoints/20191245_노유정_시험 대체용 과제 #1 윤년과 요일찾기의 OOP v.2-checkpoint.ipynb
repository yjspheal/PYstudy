```python
import datetime

class L_D:
    def __init__(self,year,month,day):
        self.year=year
        self.month=month
        self.day=day
        
    def input_date():
        tdy=datetime.datetime.now()
        print('오늘 날짜는 [%s]입니다.'%tdy.strftime('%Y-%m-%d'))
        date=input('주변사람의 다가오는 생일을 입력하세요(YYYY-MM-DD) ')
        Y,M,D=date.split('-')
        Y=int(Y); M=int(M); D=int(D)
        return Y,M,D
    
    
    def get_day_n(self):
        d=0
        for i in range (1,self.year):
            dan=L_D(i,0,0)
            if L_D.is_leap(dan)==True:
                d+=366
            else:
                d+=365
        for k in range(1,self.month):
            if  k==2:
                if L_D.is_leap(self)==True:
                    d+=29
                else:
                    d+=28
            elif k==4 or k==6 or k==9 or k==11:
                d+=30
            else:
                d+=31
                
        d+=self.day
        return d
    
    
    def is_leap(self):
        if self.year%400==0:
            return True
        elif self.year%100==0:
            return False
        elif self.year%4==0:
            return True
        else:
            return False
        
class Birthday(L_D):
    def __init__(self,year,month,day,name,age,mfm):
        L_D.__init__(self,year,month,day)
        self.name=name
        self.age=age
        self.mfm=mfm
        
    def present(age,sex):
        age=int(age/10)
        M0={'캐릭터운동화':30000,'장난감':30000}
        M1={'브랜드운동화':100000,'지갑':40000,'닌텐도스위치':300000}
        M2={'시계':200000,'명품 지갑':60000,'드론':400000}
        M3={'손수건':40000,'명품 지갑':60000,'향수':30000}
        M4={'구두':100000,'명품 지갑':60000}
        M5={'시계':200000,'드론':400000}
        FM0={'캐릭터운동화':30000,'장난감':30000,'과자':5000}
        FM1={'핸드폰케이스':100000,'다이어리':20000,'에어팟':200000}
        FM2={'향수':30000,'옷':60000}
        FM3={'목걸이':70000,'가방':300000,'향수':50000}
        FM4={'구두':100000,'명품지갑':400000}
        FM5={'시계':200000,'스카프':60000}
        lst=[M0,M1,M2,M3,M4,M5,FM0,FM1,FM2,FM3,FM4,FM5]
        if sex=='여성':
            Match=lst[6+age]
            for k in Match.keys():
                print(k,end=' ')
            print()
        else:
            Match=lst[age]
            for k in Match.keys():
                print(k,end=' ')
            print()
        
        while True:
            respon1=input('추천 선물을 입력하시면 가격을 표시합니다.\n더이상 프로그램 실행을 원하지 않는다면 0을 눌러 종료할 수 있습니다.')
            if respon1=='0':
                break
            else:
                print('%s의 가격은 평균 %d원입니다.'%(respon1,Match[respon1]))
            
    def checker(self):
        dL=['일','월','화','수','목','금','토']
        tdy=datetime.datetime.now()
        Tdy=L_D(tdy.year,tdy.month,tdy.day)
        Bday=L_D.get_day_n(self)
        Tday=L_D.get_day_n(Tdy)
        print(self.name+'님의 생일은',dL[Bday%7]+'요일 입니다.')
        print(self.name+'님의 생일까지',Bday-Tday,'일 남았습니다.')
        AD=int(self.age/10)*10
        if self.age<10:
            respon=input('%s %s이 받고싶어하는 생일 선물 리스트를 보시겠습니까?(네/아니오)'%('10대 미만',self.mfm))
        elif self.age<50:
            respon=input('%d대 %s이 받고싶어하는 생일 선물 리스트를 보시겠습니까?(네/아니오)'%(AD,self.mfm))
        else:
            respon=input('%s %s이 받고싶어하는 생일 선물 리스트를 보시겠습니까?(네/아니오)'%('50대 이상',self.mfm))
        if respon=='네':
            Birthday.present(AD,self.mfm)
        else:
            print('선물 순위를 보지 않습니다')
        
if __name__=="__main__":
    Y,M,D=L_D.input_date()   
    N,A,S=input('이 날은 누구의 생일인가요?(이름,나이,성별)').split(',')
    A=int(A)
    m=Birthday(Y,M,D,N,A,S)
    m.checker()
```
